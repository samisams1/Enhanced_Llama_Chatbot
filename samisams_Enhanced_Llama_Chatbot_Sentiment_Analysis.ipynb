```python
# Enhanced Llama Chatbot with Sentiment Analysis

This project enhances a Llama chatbot by integrating a sentiment analysis model from Hugging Face. The chatbot is designed to respond dynamically to user inputs based on the detected sentiment—positive, negative, or neutral. By analyzing the emotional tone of user messages, the chatbot can provide tailored responses that improve user interaction. This functionality is particularly useful in various contexts, such as customer support, mental health assistance, and educational tutoring. The goal is to create a more engaging and responsive chatbot that effectively addresses user needs and emotions.
```

```python
from transformers import pipeline

# Load sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
def chatbot_response(user_input):
    sentiment = sentiment_pipeline(user_input)[0]
    label = sentiment['label']
    score = sentiment['score']

    if label == 'NEGATIVE':
        return "I'm sorry to hear that. How can I assist you?"
    elif label == 'POSITIVE':
        return "That's great to hear! How can I help you further?"
    else:
        return "Thank you for your input! What would you like to know?"
```

```python
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break
    response = chatbot_response(user_input)
    print("Chatbot:", response)
```
